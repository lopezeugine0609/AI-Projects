# Welding Defects CNN Analysis

This notebook explores synthetic weld defect image generation, model architecture, and training behavior for the `ai_welding_defects` project.

```python
import matplotlib.pyplot as plt
import numpy as np
from ai_welding_defects.src.data import generate_synthetic_weld_images, CLASS_NAMES

images, labels = generate_synthetic_weld_images(num_samples=12, image_size=(64, 64))
fig, axes = plt.subplots(3, 4, figsize=(12, 8))
for ax, image, label in zip(axes.flatten(), images, labels):
    ax.imshow(image.squeeze(), cmap='gray')
    ax.set_title(CLASS_NAMES[label])
    ax.axis('off')
plt.tight_layout()
```

```python
from ai_welding_defects.src.model import build_cnn_model

model = build_cnn_model(input_shape=(64, 64, 1), num_classes=len(CLASS_NAMES))
model.summary()
```
